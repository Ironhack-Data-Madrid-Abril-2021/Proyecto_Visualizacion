¿Cómo podríamos crear el torneo más potetente de League of Legends que reventáse el mundo de los Esports?


Con en incremento del seguimiento en España de los Esports, hemos decidido que es el momento de crear un torneo internacional en Madrid.


Los participantes de este torneo tienen que ser los mejores a nivel mundial, pero tambien es muy importante que sean partidas espectáculares.



El XtremeLoL se celebrará en Octubre de 2021, necesitamos eficiencia a la hora de definir los 16 equipos que se enfrentarán en un torneo eliminatorio.


Para esta tarea, hemos designado al Data Specialist Vicente Díaz. 


De nuestra DB, necesitamos los insights suficientes para definir a todos los equipos de las distintas ligas.



¿Cuales son los equipos con la mejor Ratio de victoria/derrota?

¿Qué equipos tienen la mejor ratio de Kills/Deaths/Assists?

¿Qué equipos son los que tienen mayor cantidad de asesinatos múltiples en  los datos?



Para este proceso, hemos utilizado Pandas para la ETL.

El DataFrame constaba de más de 400 columnas, por lo que hemos definido tres nuevos DataFrames donde escogíamos las columnas válidas para cada una de las ocho mejores ligas que hay en el mundo:




-LPL: League of Legends Pro League es la liga profesional de primer nivel de League of Legends en China

-LEC: la liga profesional de deportes de League of Legends dirigida por Riot Games

-LCS: La League Championship Series es la máxima competición profesional de League of Legends en Estados Unidos y Canadá. 

-LCK: League of Legends Champions Korea es la competencia principal para los deportes electrónicos de League of Legends en Corea del Sur.

-TCL: La liga profesional de League of Legends de Turquía.

-CBLOL: El Campeonato Brasileiro de League of Legends es el nivel más alto de la competencia profesional League of Legends en  Brasil.

-LCL: League of Legends Continental League es el nivel más alto de la competencia profesional League of Legends en la Comunidad de Estados Independientes, organizada por Riot Games Rusia

-PCS: La Pacific Championship Series es una liga profesional de deportes electrónicos para jugadores de League of Legends que compiten en Taiwán, Hong Kong, Macao y el sudeste asiático 



Cabe destacar, que los equipos cuando compiten, pueden jugar en el lado azul (t_1) o en el lado rojo(t_2).

Aún así todos los equipos han de jugar en ambas posiciones para que sea justo, ya que dependiendo de la posición, tienen ventajas en función de la composición del equipo. Por lo que el análisis tiene en cuenta el resultado obtenido por el equipo en ambas situaciones.



El primer DataFrame se define para sacar las métricas del Winrate(%victorias/derrotas) de los equipos pertenecientes a estas ligas. Como filtro, hemos seleccionado a aquellos equipos que tengan un winrate acumulado en el lado azul y el lado rojo de, como mínimo un 50%.



Tras este filtrado, se determinaron 25 equipos que cumplieran este requisito.




El segundo DataFrame para detectar qué equipos de las mejores ligas profesionales tienen un KDA (Kill/Death/Assist). El requisito para este DF es que los equipos realizaran mas de 15 kills y más de 35 assists de media por partido.



Tras este proceso bastante mas complicado de pasar, sólo cuatro equipos pasaron la criba:

-Royal never give up
-Vorax
-Mad lions
-PSG Talon
-Supermassive Esports






El tercer DataFrame es sólo para analizar qué equipos tienen mayor potencial de realizar asesinatos múltiples, desde Doublekills hasta Pentakills. Una Pentakill consiste en que un solo jugador del equipo, asesine a todo el equipo rival, algo muy difícil de conseguir y más en competiciovecesidad de pasar alguna de las cribas anteriores.
