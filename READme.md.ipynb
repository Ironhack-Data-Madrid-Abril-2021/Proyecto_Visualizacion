Cmo podramos crear el torneo ms potetente de League of Legends que reventse el mundo de los Esports?


Con en incremento del seguimiento en Espaa de los Esports, hemos decidido que es el momento de crear un torneo internacional en Madrid.


Los participantes de este torneo tienen que ser los mejores a nivel mundial, pero tambien es muy importante que sean partidas espectculares.



El XtremeLoL se celebrar en Octubre de 2021, necesitamos eficiencia a la hora de definir los 16 equipos que se enfrentarn en un torneo eliminatorio.


Para esta tarea, hemos designado al Data Specialist Vicente Daz. 


De nuestra DB, necesitamos los insights suficientes para definir a todos los equipos de las distintas ligas.



Cuales son los equipos con la mejor Ratio de victoria/derrota?

Qu equipos tienen la mejor ratio de Kills/Deaths/Assists?

Qu equipos son los que tienen mayor cantidad de asesinatos mltiples en  los datos?



Para este proceso, hemos utilizado Pandas para la ETL.

El DataFrame constaba de ms de 400 columnas, por lo que hemos definido tres nuevos DataFrames donde escogamos las columnas vlidas para cada una de las ocho mejores ligas que hay en el mundo:

-LEC: la liga profesional de deportes de League of Legends dirigida por Riot Games

-LCS: La League Championship Series es la mxima competicin profesional de League of Legends en Estados Unidos y Canad. 

-LCK: League of Legends Champions Korea es la competencia principal para los deportes electrnicos de League of Legends en Corea del Sur.

-TCL: La liga profesional de League of Legends de Turqua.

-CBLOL: El Campeonato Brasileiro de League of Legends es el nivel ms alto de la competencia profesional League of Legends en  Brasil.

-LCL: League of Legends Continental League es el nivel ms alto de la competencia profesional League of Legends en la Comunidad de Estados Independientes, organizada por Riot Games Rusia

-PCS: La Pacific Championship Series es una liga profesional de deportes electrnicos para jugadores de League of Legends que compiten en Taiwn, Hong Kong, Macao y el sudeste asitico 



Cabe destacar, que los equipos cuando compiten, pueden jugar en el lado azul (t_1) o en el lado rojo(t_2).

An as todos los equipos han de jugar en ambas posiciones para que sea justo, ya que dependiendo de la posicin, tienen ventajas en funcin de la composicin del equipo. Por lo que el anlisis tiene en cuenta el resultado obtenido por el equipo en ambas situaciones.



El primer DataFrame se define para sacar las mtricas del Winrate(%victorias/derrotas) de los equipos pertenecientes a estas ligas. Como filtro, hemos seleccionado a aquellos equipos que tengan un winrate acumulado en el lado azul y el lado rojo de, como mnimo un 50%.



Tras este filtrado, se determinaron 25 equipos que cumplieran este requisito.




El segundo DataFrame para detectar qu equipos de las mejores ligas profesionales tienen un KDA (Kill/Death/Assist). El requisito para este DF es que los equipos realizaran mas de 15 kills y ms de 35 assists de media por partido.

El tercer DataFrame es slo para analizar qu equipos tienen mayor potencial de realizar asesinatos mltiples, desde Doublekills hasta Pentakills. Una Pentakill consiste en que un solo jugador del equipo, asesine a todo el equipo rival, algo muy difcil de conseguir y ms en competiciovecesidad de pasar alguna de las cribas anteriores.
